**Еще один плюс использования дат в качестве индексов – возможность выбрать данные за интересующий нас промежуток времени. В переменную bikes_summer сохраните наблюдения с 1 июня по 31 августа. Затем запишите в top_destination наиболее популярный пункт назначения (его название). Агрегируйте данные по дням и определите, в какой день в полученный пункт (top_destination) было совершено меньше всего поездок. Дату сохраните в bad_day, отформатировав timestamp с помощью .strftime('%Y-%m-%d').**

**Могут пригодиться:**

 - loc
 - strftime
 - idxmin, idxmax
 - size
 - query

```python
# Выбираем по индексу интересующий нас временной период

bikes_summer = bikes['2018-06-01':'2018-08-31']

# ИЛИ bikes_summer = bikes.loc['2018-06-01':'2018-08-31']
```

```python
# Находим наиболее популярное место назначения (столбец to_station_name)

top_destination = bikes_summer.to_station_name\
                              .value_counts()\
                              .idxmax()
```

```python
top_destination
```

```python
# Фильтруем ДФ по наименованию самой популярной станции и группируем данные по дням с высчитыванием минимального значения даты

bad_day = bikes_summer.query("to_station_name == 'Streeter Dr & Grand Ave'")\
                      .resample('1d')\
                      .trip_id.count()\
                      .idxmin()\
                      .strftime('%Y-%m-%d') # Метод форматирования типа timestamp в str
```

```python
bad_day
```

```python
# Создаем дополнительную колонку с название дня недели

bikes_summer = bikes_summer.assign(weekday = lambda x: pd.to_datetime(x.index).strftime('%A'))
```

```python
bikes_summer.head(3)
```

```python
# Группируем по дню недели и пункту назначения и подсчитываем количество поездок с дальнейшим применением unstack

days = bikes_summer.groupby(['weekday','to_station_name']).agg({'trip_id':'count'}).unstack(level=0)
```

```python
days.info()
```

```python
# Формируем столбец working_days

days['working_days'] =    days[('trip_id', 'Monday')] 
                        + days['trip_id', 'Tuesday'] 
                        + days['trip_id', 'Wednesday'] 
                        + days['trip_id', 'Thursday'] 
                        + days['trip_id', 'Friday']
```

```python
# Формируем столбец days_off

days['days_off'] =    days[('trip_id', 'Saturday')]
                    + days['trip_id', 'Sunday']
```

```python
# Наиболее популярный и второй по популярности пункт назначения в субботу

days.sort_values(('trip_id', 'Saturday'), ascending = False)
```

```python
# Наиболее популярный пункт назначения в рабочие дни

days.sort_values('working_days', ascending = False)
```
