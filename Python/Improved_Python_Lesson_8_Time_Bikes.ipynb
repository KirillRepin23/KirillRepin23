**Данные за**

 - Q2: (1) апрель, (2) май, (3) июнь
 - Q3: (4) июль, (5) август, (6) сентябрь

```python
bikes_q2_sample_apr = pd.read_csv('./Data_for_time_managing_task/bikes_q2_sample_apr.csv', parse_dates=[1,2])
bikes_q2_sample_may = pd.read_csv('./Data_for_time_managing_task/bikes_q2_sample_may.csv', parse_dates=[1,2])
bikes_q2_sample_jun = pd.read_csv('./Data_for_time_managing_task/bikes_q2_sample_jun.csv', parse_dates=[1,2])
bikes_q3_sample_july = pd.read_csv('./Data_for_time_managing_task/bikes_q3_sample_july.csv', parse_dates=[1,2])
bikes_q3_sample_aug = pd.read_csv('./Data_for_time_managing_task/bikes_q3_sample_aug.csv', parse_dates=[1,2])
bikes_q3_sample_sep = pd.read_csv('./Data_for_time_managing_task/bikes_q3_sample_sep.csv', parse_dates=[1,2])
bikes_q4_sample_oct = pd.read_csv('./Data_for_time_managing_task/bikes_q4_sample_oct.csv', parse_dates=[1,2])
bikes_q4_sample_nov = pd.read_csv('./Data_for_time_managing_task/bikes_q4_sample_nov.csv', parse_dates=[1,2])
bikes_q4_sample_dec = pd.read_csv('./Data_for_time_managing_task/bikes_q4_sample_dec.csv', parse_dates=[1,2])
```

```python
bikes = pd.concat([
                 bikes_q2_sample_apr,
                 bikes_q2_sample_may,
                 bikes_q2_sample_jun,
                 bikes_q3_sample_july,
                 bikes_q3_sample_aug,
                 bikes_q3_sample_sep,
                 bikes_q4_sample_oct,
                 bikes_q4_sample_nov,
                 bikes_q4_sample_dec
                 ]
                )

# Либо можно сразу в листе указывать считывание по ссылкам на части данных:
# bikes = pd.concat([
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_apr.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_may.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q2_sample_jun.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_july.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_aug.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q3_sample_sep.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_oct.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_nov.csv'),
#                  pd.read_csv('https://stepik.org/media/attachments/lesson/367415/bikes_q4_sample_dec.csv')
#                  ]
#                 )
```

```python
# Группируем по типу пользователя и переносим Customer и Subscriber в столбцы с помощью .unstack

bikes_trips_per_usertype = bikes.groupby('usertype').resample('1d').trip_id.count().unstack(level=0)
```

```python
bikes_trips_per_usertype
```

```python
# Выбираем дни по условию Customer > Subscriber

bikes_trips_per_usertype.query('Customer > Subscriber')
```

```python
# Другой способ - через .size() и .T

br = bikes.groupby(['usertype']).resample('1d').size().T
```

```python
br
```

```python
br[br.Customer > br.Subscriber]
```
